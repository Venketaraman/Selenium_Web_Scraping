```python
# Import Webdriver, Chrome, Service, Select and Keys

from selenium import webdriver
from webbrowser import Chrome
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
```

```python
# Directing to Chrome
options = webdriver.ChromeOptions()
options.add_experimental_option("detach",True)
website = 'https://www.google.com/search?q=WHO+reveals+leading+causes+of+death+and+disability+worldwide%3A+2000-2019&rlz=1C1GCEB_enIN996IN996&ei=Z93AY8TpAf2DhbIP7NG32A4&ved=0ahUKEwiEudWS2sP8AhX9QUEAHezoDesQ4dUDCA8&uact=5&oq=WHO+reveals+leading+causes+of+death+and+disability+worldwide%3A+2000-2019&gs_lcp=Cgxnd3Mtd2l6LXNlcnAQAzoTCAAQjwEQ6gIQtAIQjAMQ5QIYAToTCC4QjwEQ6gIQtAIQjAMQ5QIYAUoECEEYAEoECEYYAFCABliABmCNCGgBcAB4AIABnwGIAZ8BkgEDMC4xmAEAoAEBoAECsAEKwAEB2gEECAEYCg&sclient=gws-wiz-serp'
path = 'C:/Users/z034660/WHO_COVID_ANALYSIS/chromedriver'
path =  'C:/Users/Venket/Documents/Python Scripts\Machine Learning/Web Scrapping/WHO_COVID_ANALYSIS'
```

```python
#Installing drive manager
driver = webdriver.Chrome (options = options, service = Service(ChromeDriverManager().install()))
driver.maximize_window()
driver.get(website)
```

```python
#Directing to the site
time.sleep(2)
directing_web_page = driver.find_element(By.XPATH, "//div[@class='g']//div//div//div[@class='kvH3mc BToiNc UK95Uc']//h3[@class='LC20lb MBeuO DKV0Md'][contains(text(),'WHO reveals leading causes of death and disability')]")
time.sleep(2)
directing_web_page.click()
time.sleep(2)
```

```python
#Auto clicking the data tab
data_tab = driver.find_element(By.XPATH, "//ul[@class='nav']//a[normalize-space()='Data']")
time.sleep(2)
data_tab.click()
time.sleep(2)
```

```python
#From drop down selecting Covid 19 dashboard
Selecting_dashboard = driver.find_element(By.XPATH, "//ul[@class='main-navigation navVertical']//a[contains(text(),'COVID-19 Dashboard')]")
time.sleep(2)
Selecting_dashboard.click()
time.sleep(7)
#Auto Scroll Down
driver.execute_script("window.scrollBy(0,800)","")
time.sleep(5)
driver.execute_script("window.scrollBy(0,1000)","")
time.sleep(5)
driver.execute_script("window.scrollBy(0,1300)","")
time.sleep(5)
driver.execute_script("window.scrollBy(0,1700)","")
time.sleep(5)
driver.execute_script("window.scrollBy(0,2100)","")
time.sleep(3)
#driver.execute_script("window.scrollBy(0,2500)","")
#time.sleep(5)
```

```python
#Select Tabel_View
Table_view = driver.find_element(By.XPATH, "//li[@class='sc-fzplgP cYAPyO']//a[normalize-space()='Table View']")
time.sleep(3)
Table_view.click()
time.sleep(3)
#Auto Scroll Down
scroll = driver.find_element(By.XPATH, "//div[@class='sc-AxjAm ekIhFm']//a[normalize-space()='Table View']")
driver.execute_script("window.scrollBy(0,300)","")
time.sleep(3)
driver.execute_script("window.scrollBy(0,350)","")
time.sleep(3)
#driver.execute_script("window.scrollBy(0,450)","")
#time.sleep(2)
#Go to top
go_top = driver.find_element(By.XPATH, "//div[@class='sc-AxjAm ekIhFm']//span[contains(text(),'Back to top')]")
time.sleep(3)
go_top.click()
time.sleep(1)
```

```python
#Fetching more resources
resources = driver.find_element(By.XPATH, "//div[@class='sc-AxjAm ekIhFm']//a[normalize-space()='More Resources']")
time.sleep(3)
resources.click()
time.sleep(3)
#Auto Scroll Down
driver.execute_script("window.scrollBy(0,800)","")
time.sleep(3)
driver.execute_script("window.scrollBy(0,1100)","")
time.sleep(2)
#Go to top
top1 = driver.find_element(By.XPATH, "//div[@class='sc-AxjAm ekIhFm']//span[contains(text(),'Back to top')]")
time.sleep(2)
top1.click()
time.sleep(1)
```

```python
#Select Data
data1 = driver.find_element(By.XPATH, "//div[@class='sc-AxjAm ekIhFm']//a[normalize-space()='Data']")
time.sleep(3)
data1.click()
time.sleep(3)
#Auto Scroll Down
driver.execute_script("window.scrollBy(0,800)","")
time.sleep(3)
driver.execute_script("window.scrollBy(0,1500)","")
time.sleep(3)
#Download Global data and vaccine data
Globa_data = driver.find_element(By.XPATH, "//div[@id='Data Download']//div[@class='sc-AxjAm cMSDjZ']")
time.sleep(2)
Globa_data.click()
vaccine_data = driver.find_element(By.XPATH, "//a[contains(text(),'https://covid19.who.int/WHO-COVID-19-global-table-')]")
time.sleep(4)
vaccine_data.click()
#Auto Scroll Down
driver.execute_script("window.scrollBy(0,700)","")
time.sleep(2)
```
